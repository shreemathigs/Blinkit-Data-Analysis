## While Loop

```python
#While Loop
  #Everything should be manual
  #Assign the condition and also stop the iteration

#Syntax:

#Intialization of Variable

while <condition>:
  <code>
  iteration
```

## Nested if Statement

```python
#Nested

age= int(input('Enter your age:'))
test = input('Test Status(Pass/Fail):')

if ( age>=18):
  print('Allowed')
  if test == 'pass' or 'Pass':
    print('Grant License')
elif (age==17):
  print('Waiting List')
else:
  print('Not Allowed')
```

```python
#Loan Process

# sal > 5000
# cibil > 700
# loan_amount < 10000
# Dispose Loan


sal=int(input('Enter Sal:'))
cibil=int(input('Enter cibil:'))
loan=int(input('Enter loan amount:'))

if sal >5000:
  print('Salary Check Passed !!')

  if cibil > 700:
      print('Cibil Check Passed !!')

      if loan <10000:
          print('Loan Granted !!')
      else:
          print('Limit Exceeded !!')

  else:
      print('Cibil Check Failed !!')

else:
  print('Salary Check Failed !!')
```

## Multiple Conditions Handling

```python
#Multiple conditions

#AND
#OR


#AND--
<condition> and <conditon>

T , T --> True
T , F --> False
F , T --> False
F , F --> False

#OR--
<condition> or <conditon>

T , T --> True
T , F --> True
F , T --> True
F , F --> False
```

```python
#Checking of age between 30 to 40 --- AND

age = int(input('Enter age:'))

if (age >= 30) and (age <= 40):
  print('Age Granted')
else:
  print('Out of range')
```

```python
# Using of OR
age = int(input('Enter age:'))

if (age == 30) or (age == 40):
  print('Age Granted')
else:
  print('Not Allowed')
```
